O período considerado foi entre 2009 e 2020, com janelas de análise de 5 e 10 anos. Os ativos foram selecionados com base no market cap atual e tempo de mercado do ativo. Os ETFs VTI, IJR, SHV, TLT e GLD foram escolhidos conforme sugestão na página sobre o Golden Butterfly no site <a href="http://www.lazyportfolioetf.com/allocation/golden-butterfly/">Lazy Portfolio ETF</a> e sugestões do analista Otávio Paranhos. 

Taxas, custos de transação e dividendos não foram considerados. 

**Metogologia de rebalanceamento:** A carteira foi rebalanceada no primeiro pregão de cada trimestre. Valores em moeda estrangeira foram convertidos para real pela taxa de conversão de fechamento do dia. 

```python
import psutil
import datetime
import time
import pandas as pd
import numpy as np
import yfinance as yf
import yahooquery as yq
import matplotlib.pyplot as plt

###################################### HELPER FUNCTIONS ######################################
def get_ticker_history(ticker, start_date, end_date=None):
    """
    Fetch Yahoo finance to get quotation history of a given ticker
    -----------------------------------------------------------------------------------------
    Input:
    ticker: ticker of the asset
    start_date: date from which to fetch the history
    end_date: last date to fetch (if none, today is used)
    -----------------------------------------------------------------------------------------
    Return:
    Prices of the ticker or None
    """
    if end_date == None:
        end_date = datetime.date.today().strftime("%Y-%m-%d")
    ticker_obj = yq.Ticker(ticker)
    try:
        ticker_history = ticker_obj.history(start=start_date, end=end_date)
        ticker_history.reset_index(inplace=True)
        return ticker_history
    except:
        print("Error on fetching history of",ticker)
        
def date_conversion_br_bonds(date):
    """
    Convert date strings of formats dd/mm/yyyy and yyyy-mm-dd to timestamp
    -----------------------------------------------------------------------------------------
    Input:
    date: date to be converted (either timestamp or string)
    -----------------------------------------------------------------------------------------
    Return:
    Date converted to timestamp
    """
    if type(date)==type(""):
        if "-" in date:
            return pd.to_datetime(date, format="%Y-%m-%d")
        elif "/" in date:
            return pd.to_datetime(date, format="%d/%m/%Y")
    else:
        return date

def start_end_dates_between_n_months(df, n_months, last_date=None):
    """
    Return the first date after a given last_date and the last valid date after n_months.
    Assumes the date is the index of df
    -----------------------------------------------------------------------------------------
    Input:
    df: dataframe in which the valid dates are represented as its index
    n_months: number of months to shift
    last_date: date from which to start the algorithm, if null the minimum date of df is used
    -----------------------------------------------------------------------------------------
    Return:
    (start, end): tuple of the first and last valid dates within last_date + n_months
    """
    if last_date==None:
        initial_date = df.index.min()
    else:
        initial_date = df[df.index>last_date].index.min()
        
    theoretic_end = initial_date + pd.tseries.offsets.MonthEnd(n_months)
    end_date = df[(df.index.year==theoretic_end.year) & (df.index.month==theoretic_end.month)].index.max()
    return (initial_date, end_date)

def define_rebalancing_dates(df, n_months):
    """
    Define the dates to rebalance the portfolio considering it will happen every n_months
    -----------------------------------------------------------------------------------------
    Input:
    df: dataframe in which the valid dates are represented as its index
    n_months: number of months to shift
    -----------------------------------------------------------------------------------------
    Return:
    date_intervals: array of (first_day, last_day), where the portfolio must be rebalanced 
    on first_day
    """
    date_intervals=[]
    start_date=df.index.min() - datetime.timedelta(days = 1)
    while(start_date<df.index.max()):
        s, e = start_end_dates_between_n_months(df, 3, start_date)
        date_intervals.append((s, e))
        start_date = e
    return date_intervals

def simulation(df, allocation, rebalancing_dates, initial_budget=10000):
    """
    Perform asset allocation and rebalancing according to the allocation and 
    rebalancing_dates provided
    -----------------------------------------------------------------------------------------
    Input:
    df: dataframe in which the valid dates are represented as its index
    allocation: ideal allocation for each asset in portfolio
    rebalancing_dates: list of tuples indicating when to rebalance the portfolio
    initial_budget: amount of money invested. changing this will not influence the total return
    -----------------------------------------------------------------------------------------
    Return:
    index: first year of the moving window
    total_return: total  return on this timeframe (in percentage)
    qty_history: quantity of each asset after every portfolio rebalancing
    """
    budget = initial_budget
    qty_history=[]
    for (start, end) in rebalancing_dates:
        qty = {}
        new_budget = 0
        for asset in allocation.keys():
            qty[asset] = budget*allocation[asset]/df[df.index==start][asset].values[0]
            new_budget += qty[asset]*all_assets[df.index==end][asset].values[0]
        qty_history.append({start: qty})
        budget = new_budget
        
    index = rebalancing_dates[0][0].year
    total_return = budget/initial_budget - 1
    return index, total_return, qty_history
```

```python
#Get data from Yahoo Finance
th = {}
for ticker in ['USDBRL=X', 'BOVA11.SA', 'SMAL11.SA', 'VTI', 'IJR', 'SHV', 'TLT', 'GDL', 'GDX']:
    th[ticker] = get_ticker_history(ticker, "2009-01-02", "2020-12-31")[['date', 'adjclose']]

#Structure the data as a dataframe
try:
    del tickers_all
except:
    pass

for ticker in th.keys():
    th[ticker].columns = ['date', ticker]
    try:
        tickers_all = tickers_all.merge(th[ticker], left_on=['date'], right_on=['date'], how='outer')
    except:
        tickers_all = th[ticker]

tickers_all.sort_values(by='date', inplace=True)
tickers_all.fillna(method='ffill', inplace=True) #Fill nulls with last valid price
tickers_all.set_index('date', inplace=True)

# Conversion from USD to BRL
tickers_brl = tickers_all.copy()
for col in ['VTI', 'IJR', 'SHV', 'TLT', 'GDL', 'GDX']:
    tickers_brl[col] = tickers_brl[col]*tickers_brl['USDBRL=X']
```

```python
#Get brazilian bonds data
ntnb = pd.read_excel('../data/ntnb_principal.ods', engine='odf')[['Dia', 'PU Extrato 9:00']]
lft = pd.read_excel('../data/lft.ods', engine='odf')[['Dia', 'PU Extrato 9:00']]

#Preprocessing
ntnb.columns = ['Dia', 'NTNB']
lft.columns = ['Dia', 'LFT']
ntnb['Dia'] = ntnb['Dia'].apply(lambda x: date_conversion_br_bonds(x))
lft['Dia'] = lft['Dia'].apply(lambda x: date_conversion_br_bonds(x))
ntnb.sort_values(by='Dia', inplace=True)
lft.sort_values(by='Dia', inplace=True)
ntnb.set_index('Dia', inplace=True)
lft.set_index('Dia', inplace=True)
```

```python
#Consolidate all assets in a single dataframe
all_assets = tickers_brl.copy()
all_assets = all_assets.merge(ntnb, left_index=True, right_index=True, how='outer')
all_assets = all_assets.merge(lft, left_index=True, right_index=True, how='outer')
all_assets.fillna(method='ffill', inplace=True) #Fill nulls with last valid price
```

```python
#Here starts the backtest itself. First the allocation strategies are set up, 
#then the algorithm compute the backtests for every timeframe and portfolio.

#Portfolios setup
allocations = {'Golden Butterfly Brasil': {'BOVA11.SA': 0.2,              #Total market
                                           'SMAL11.SA': 0.2,              #Small caps
                                           'LFT':0.2,                     #Short term bonds
                                           'NTNB': 0.2,                   #Long term bonds
                                           'GDL': 0.2},                   #Gold
               
               'Golden Butterfly BR+US':  {'BOVA11.SA': 0.1, 'VTI': 0.1,  #Total market
                                           'SMAL11.SA': 0.1, 'IJR': 0.1,  #Small caps
                                           'LFT':0.1, 'SHV': 0.1,         #Short term bonds
                                           'NTNB': 0.1, 'TLT': 0.1,       #Long term bonds
                                           'GDL': 0.2},                   #Gold
               
               'Miner Butterfly BR+US':   {'BOVA11.SA': 0.1, 'VTI': 0.1,  #Total market
                                           'SMAL11.SA': 0.1, 'IJR': 0.1,  #Small caps
                                           'LFT':0.1, 'SHV': 0.1,         #Short term bonds
                                           'NTNB': 0.1, 'TLT': 0.1,       #Long term bonds
                                           'GDL': 0.1, 'GDX': 0.1}}       #Gold and mining
```

```python
#Portfolios will be rebalanced every 3 months
n_months = 3
size_1yr = int(12/n_months)
size_5yr = 5*size_1yr
size_10yr = 10*size_1yr
rebalancing_dates = define_rebalancing_dates(all_assets, n_months)

#Here the slices of rebalancing dates for each moving window are defined
windows_5yr = [rebalancing_dates[x:x+size_5yr] for x in range(0, len(rebalancing_dates)-size_5yr+1, size_1yr)]
windows_10yr = [rebalancing_dates[x:x+size_10yr] for x in range(0, len(rebalancing_dates)-size_10yr+1, size_1yr)]

strategies_simulations = {} 
for strategy in allocations.keys():
    allocation = allocations[strategy]
    
    #The simulation - 5 yr window
    allocations_5yr = {}
    returns_5yr = {}
    for rebalancing_dates_window in windows_5yr:
        index, total_return, qty_history = simulation(all_assets, allocation, rebalancing_dates_window)
        allocations_5yr[index] = qty_history
        returns_5yr[index] = total_return

    #The simulation - 10 yr window
    allocations_10yr = {}
    returns_10yr = {}
    for rebalancing_dates_window in windows_10yr:
        index, total_return, qty_history = simulation(all_assets, allocation, rebalancing_dates_window)
        allocations_10yr[index] = qty_history
        returns_10yr[index] = total_return    
        
    #The simulation - whole period
    allocations_tot = {}
    returns_tot = {}
    index, total_return, qty_history = simulation(all_assets, allocation, rebalancing_dates)
    allocations_tot[index] = qty_history
    returns_tot[index] = total_return 
        
    strategies_simulations[strategy] = {'returns_5yr': returns_5yr, 
                                        'allocations_5yr': allocations_5yr,
                                        'returns_10yr': returns_10yr,
                                        'allocations_10yr': allocations_10yr,
                                        'returns_tot': returns_tot,
                                        'allocations_tot': allocations_tot}
```

```python
returns_5yr = [(x,strategies_simulations[x]['returns_5yr']) for x in strategies_simulations.keys()]
returns_10yr = [(x,strategies_simulations[x]['returns_10yr']) for x in strategies_simulations.keys()]
returns_tot = [(x,strategies_simulations[x]['returns_tot']) for x in strategies_simulations.keys()]
```

```python
returns_dfs = []
for i in range(len(returns_5yr)):
    df = pd.DataFrame([returns_5yr[i][1]], index=[returns_5yr[i][0]]).T
    returns_dfs.append(df)
returns_5yr_df = pd.concat(returns_dfs,axis=1)
returns_5yr_df.columns = pd.MultiIndex.from_arrays([["5yr"]*3,returns_5yr_df.columns])

returns_dfs = []
for i in range(len(returns_10yr)):
    df = pd.DataFrame([returns_10yr[i][1]], index=[returns_10yr[i][0]]).T
    returns_dfs.append(df)
returns_10yr_df = pd.concat(returns_dfs,axis=1)
returns_10yr_df.columns = pd.MultiIndex.from_arrays([["10yr"]*3,returns_10yr_df.columns])

returns_dfs = []
for i in range(len(returns_tot)):
    df = pd.DataFrame([returns_tot[i][1]], index=[returns_tot[i][0]]).T
    returns_dfs.append(df)
returns_tot_df = pd.concat(returns_dfs,axis=1)
returns_tot_df.columns = pd.MultiIndex.from_arrays([["All"]*3,returns_tot_df.columns])

returns_all_df = pd.concat([returns_5yr_df, returns_10yr_df], axis=1)


#Calculate grand mean and append it to returns_all_df
mean = returns_all_df.mean().to_frame()
mean.columns = ['Média']
mean = mean.T
returns_all_df = pd.concat([returns_all_df, mean], axis=0)
```

A tabela abaixo apresenta o desempenho das estratégias avaliadas em janelas móveis de 5 e 10 anos. O portfólio "Golden Butterfly Brasil" perde em todas as comparações. **O "Golden Butterfly BR+US" ganha em 6 das 8 janelas moveis de 5 anos e em todas as janelas de 10 anos.**

```python
qty_assets = []
for alloc in strategies_simulations['Golden Butterfly BR+US']['allocations_tot'][2009]:
    qty_assets.append(pd.DataFrame(alloc).T)
qty_assets_df = pd.concat(qty_assets, axis=0)
```

```python
all_assets_with_backtest = all_assets.merge(qty_assets_df, 
                                            left_index=True, 
                                            right_index=True, 
                                            how='left', 
                                            suffixes=('','_qty'))
all_assets_with_backtest.fillna(method='ffill', inplace=True) #Fill nulls with last valid allocation

assets = ['BOVA11.SA', 'SMAL11.SA', 'VTI', 'IJR', 'SHV', 'TLT', 'GDL', 'NTNB', 'LFT']
assets_qty = [x+"_qty" for x in assets]

all_assets_with_backtest['Golden Butterfly BR+US'] = (all_assets_with_backtest[assets]*all_assets_with_backtest[assets_qty].values).sum(axis=1)
all_assets_with_backtest.drop(assets_qty, axis=1, inplace=True)
```

Comparando a evolução das bolsas brasileira e americana contra o Golden Butterfly BR+US (figura abaixo), percebemos que a rentabilidade do mesmo foi intermediári, porém com um comportamento com baixa volatilidade.

```python
#Plot assets time series
cols_to_plot = ['BOVA11.SA', 'SMAL11.SA', 'VTI', 'IJR', 'Golden Butterfly BR+US']
color_dict = {'BOVA11.SA': '#1EDC00', 
              'SMAL11.SA': '#4655F5',
              'VTI': '#FF3C28',
              'IJR': '#0AB9E6',
              'Golden Butterfly BR+US': '#FAA005'}
(all_assets_with_backtest[cols_to_plot]
 /all_assets_with_backtest[cols_to_plot].iloc[0]-1).plot(figsize=(15,8),
                                                         color=[color_dict.get(x, '#333333') for x in cols_to_plot])
plt.title("Variação % do preço do Golden Butterfly BR+US e bolsas")
plt.savefig("variacao_percentual_ativos_e_portfolio.png", bbox_inches='tight', pad_inches=0.1)
plt.plot()
```

A figura a seguir apresenta o desvio padrão das diferenças diárias de cada ativo. Este indicador é uma medida razoável de volatilidade. Sob esta perspectiva, o Golden Butterfly BR+US tem volatilidade intermediária entre o Ibovespa e a NTNB.

```python
vol_assets = (all_assets_with_backtest/all_assets_with_backtest.iloc[0]).diff(1).std()
vol_assets.sort_values(ascending=True).plot(kind='bar', figsize=(15,5))
plt.title("Desvio padrão das diferenças diárias na valorização de cada ativo (normalizado)")
plt.savefig("desvio_padrao_diferencas.png", bbox_inches='tight', pad_inches=0.1)
plt.plot()
```

E apesar da baixa volatilidade, seu CAGR (taxa de crescimento anual composta) é a terceira mais alta entre todos os ativos considerados.

```python
number_years = len(all_assets_with_backtest.index.year.unique())
((all_assets_with_backtest.iloc[-1]
  /all_assets_with_backtest.iloc[0]-1)
 **(1/number_years)-1).sort_values(ascending=False).plot(kind='bar', figsize=(15,5))
plt.title("CAGR por ativo entre 2009 e 2020")
plt.savefig("cagr_por_ativo.png", bbox_inches='tight', pad_inches=0.1)
plt.plot()
```

Em resumo, conclui-se que no período analisado uma estratégia de alocação em multiclasse de ativos, diversificada entre Brasil e EUA e com rebalanceamento trimestral apresentou rentabilidade superior a maior parte dos ativos individuais do portfólio e com baixa volatilidade (comportamento _bond-like_ ).
